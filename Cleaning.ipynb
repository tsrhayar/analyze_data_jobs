```python
import re

def convert_to_numeric(salary_str):
    numbers = re.findall(r'\d+', str(salary_str))  # Convert to string to handle non-string values

    if not numbers:
        return float('nan')

    numeric_value = int(numbers[0])
    if 'K' in str(salary_str):
        numeric_value *= 1000
    if 'GBP' in str(salary_str):
        numeric_value *= 1,28
    if 'EUR' in str(salary_str):
        numeric_value *= 1,11

    return numeric_value

df.loc[:, 'Salary USD'] = df['Salary'].apply(convert_to_numeric)
df["Salary USD"] = pd.to_numeric(df["Salary USD"], errors='coerce')
```

```python
def get_country(location):
    full_address = ''
    location = location.lower()
    if 'remote' in location.lower():
        locationWithoutRemote = remove_remote_from_location(remote_list, location)
        
    exist  = False
    exist2 = False
    exist3 = False
    
    for country in countries:
        country = country.lower()
        if country in location:
            exist = True
            print("1 - location:", location, "full_address:", full_address, " / ", "country", country)
            return country
            break
    
    if not exist:
        locationWithoutRemote = remove_remote_from_location(remote_list, location)
        full_address = get_full_address(locationWithoutRemote)
        for country in countries:
            country = country.lower()
            full_address = full_address.lower()
            if country in full_address:
                print("2 - location:", location, "full_address:", full_address, " / ", "country", country)
                exist2 = True
                return country
                
                break
                
        if not exist2:
            if ";" in locationWithoutRemote:
                locationWithoutRemote = locationWithoutRemote.split(";")[0]
            elif "," in locationWithoutRemote:
                locationWithoutRemote = locationWithoutRemote.split(",")[0]
            elif "/" in locationWithoutRemote:
                location = locationWithoutRemote.split("/")[0]
            elif "(" in locationWithoutRemote:
                locationWithoutRemote = locationWithoutRemote.split("()")[0]
#                 ------------

            full_address = get_full_address(locationWithoutRemote)
            full_address = full_address.lower()
            for country in countries:
                country = country.lower()
                if country in full_address:
                    print("3 - location:", location, "full_address:", full_address, " / ", "country", country)
                    exist3 = True
                    return country

                    break

            if not exist3:
                location_lower = location.lower()
                if 'remote' in location_lower or 'home' in location_lower:
                    print("4 - Remote", "location:", location)
                    return 'Remote'
                else:
                    full_address = "Not found => ( " + location + " )"
                    print("5 - location:", location, "full_address:", full_address)
                    return full_address

# df = df.head(100)

df['Country'] = df['Location'].apply(get_country)

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def get_data_ai_job_titles():
    # Add your list of data and AI-related job titles here
    data_ai_job_titles = [
        "Data Scientist",
        "Machine Learning Engineer",
        "Artificial Intelligence (AI) Engineer",
        "Data Engineer",
        "Data Analyst",
        "Business Intelligence (BI) Analyst",
        "Big Data Engineer",
        "AI Research Scientist",
        "AI/ML Developer",
        "Data Architect",
        "Deep Learning Engineer",
        "Natural Language Processing (NLP) Engineer",
        "Computer Vision Engineer",
        "Data Warehouse Specialist",
        "Data Visualization Specialist",
        "AI Consultant",
        "Data Mining Specialist",
        "Predictive Analytics Specialist",
        "Statistical Analyst",
        "Cloud Data Engineer",
        "AI Ethics Specialist",
        "Algorithm Developer",
        "AI Product Manager",
        "Data Governance Manager",
        "Data Quality Analyst",
        "Machine Learning Operations (MLOps) Engineer",
        "AI Solutions Architect",
        "Data Science Manager",
        "AI Project Manager",
        "Data Integration Specialist",
    ]
    return data_ai_job_titles

# Get the list of data and AI job titles
data_ai_job_titles = get_data_ai_job_titles()

# Function to map job titles using TF-IDF and cosine similarity
def map_job_title(title):
    vectorizer = TfidfVectorizer()
    titles = data_ai_job_titles + [title]
    tfidf_matrix = vectorizer.fit_transform(titles)
    similarity = cosine_similarity(tfidf_matrix)
    closest_idx = similarity[-1, :-1].argmax()
    return data_ai_job_titles[closest_idx]

# Apply the mapping function to create the new "JobTitle" column
df["JobTitle"] = df["Job Title"].apply(map_job_title)

# Print the updated dataframe
df
```
