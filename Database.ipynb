```python
import pandas as pd
import ast
import pandas as pd

# Chargement des données à partir du fichier CSV dans une DataFrame
df = pd.read_csv('./clean_data_with_the_countries2.csv', encoding='ISO-8859-1')

# Renommer la colonne "Requirment of the company" en "Requirment of the company" pour corriger la faute de frappe
df = df.rename(columns={"Requirment of the company ": "Requirment of the company"})

# Convertir la colonne "Requirment of the company" de la représentation de chaîne en liste réelle
df['Requirment of the company'] = df['Requirment of the company'].apply(ast.literal_eval)

# Convertir la colonne "Facilities" de la représentation de chaîne en liste réelle
df['Facilities'] = df['Facilities'].apply(ast.literal_eval)

df = df.dropna(subset=['Country'])
df = df[~df['Country'].str.contains('not found', case=False)]
df['Salary USD'] = df['Salary USD'].fillna(0)
df['Experience level'] = df['Experience level'].fillna('not specified')
df['Job Type'] = df['Job Type'].fillna('not specified')
df['ID'] = range(1, len(df) + 1)
```

```python
import mysql.connector

# MySQL database connection details
host = 'localhost'
user = 'root'
password = ''
database = 'data_jobs'

# Function to create the database and tables
def create_tables():
    # Connect to MySQL server
    conn = mysql.connector.connect(host=host, user=user, password=password)
    cursor = conn.cursor()

    # Create the database if it doesn't exist
    cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
    cursor.execute(f"USE {database}")

    # Create the tables
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS countries (
            id INT PRIMARY KEY AUTO_INCREMENT, 
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS experience_levels (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS requirements (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS facilities (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)
    
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS job_types (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS jobs (
            id INT PRIMARY KEY,
            jobTitle VARCHAR(255) NOT NULL,
            jobDescription VARCHAR(255) NOT NULL,
            location VARCHAR(255) NOT NULL,
            salaryUSD FLOAT,
            CountryId INT DEFAULT NULL,
            JobTypeId INT DEFAULT NULL,
            ExperienceLevelId INT DEFAULT NULL,
            FOREIGN KEY (CountryId) REFERENCES countries(id),
            FOREIGN KEY (JobTypeId) REFERENCES job_types(id),
            FOREIGN KEY (ExperienceLevelId) REFERENCES experience_levels(id)
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS job_requirements (
            jobId INT,
            requirementId INT,
            FOREIGN KEY (jobId) REFERENCES jobs(id),
            FOREIGN KEY (requirementId) REFERENCES requirements(id),
            PRIMARY KEY (jobId, requirementId)
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS job_facilities (
            jobId INT,
            facilityId INT,
            FOREIGN KEY (jobId) REFERENCES jobs(id),
            FOREIGN KEY (facilityId) REFERENCES facilities(id),
            PRIMARY KEY (jobId, facilityId)
        )
    """)

    # Commit and close the connection
    conn.commit()
    conn.close()
```
