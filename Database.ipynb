```python
import pandas as pd
import ast
import pandas as pd

# Chargement des données à partir du fichier CSV dans une DataFrame
df = pd.read_csv('./clean_data_with_the_countries2.csv', encoding='ISO-8859-1')

# Renommer la colonne "Requirment of the company" en "Requirment of the company" pour corriger la faute de frappe
df = df.rename(columns={"Requirment of the company ": "Requirment of the company"})

# Convertir la colonne "Requirment of the company" de la représentation de chaîne en liste réelle
df['Requirment of the company'] = df['Requirment of the company'].apply(ast.literal_eval)

# Convertir la colonne "Facilities" de la représentation de chaîne en liste réelle
df['Facilities'] = df['Facilities'].apply(ast.literal_eval)

df = df.dropna(subset=['Country'])
df = df[~df['Country'].str.contains('not found', case=False)]
df = df.where(pd.notna(df), 0)
```

```python
import mysql.connector

# MySQL database connection details
host = 'localhost'
user = 'root'
password = ''
database = 'data_jobs'

# Function to create the database and tables
def create_tables():
    # Connect to MySQL server
    conn = mysql.connector.connect(host=host, user=user, password=password)
    cursor = conn.cursor()

    # Create the database if it doesn't exist
    cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database}")
    cursor.execute(f"USE {database}")

    # Create the tables
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS countries (
            id INT PRIMARY KEY AUTO_INCREMENT, 
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS experience_levels (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS requirements (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS facilities (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)
    
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS job_types (
            id INT PRIMARY KEY AUTO_INCREMENT,
            name VARCHAR(255) NOT NULL
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS jobs (
            id INT PRIMARY KEY AUTO_INCREMENT,
            jobTitle VARCHAR(255) NOT NULL,
            jobDescription VARCHAR(255) NOT NULL,
            location VARCHAR(255) NOT NULL,
            salaryUSD FLOAT,
            CountryId INT DEFAULT NULL ,
            FOREIGN KEY (CountryId) REFERENCES countries(id)
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS job_requirements (
            jobId INT,
            requirementId INT,
            FOREIGN KEY (jobId) REFERENCES jobs(id),
            FOREIGN KEY (requirementId) REFERENCES requirements(id),
            PRIMARY KEY (jobId, requirementId)
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS job_facilities (
            jobId INT,
            facilityId INT,
            FOREIGN KEY (jobId) REFERENCES jobs(id),
            FOREIGN KEY (facilityId) REFERENCES facilities(id),
            PRIMARY KEY (jobId, facilityId)
        )
    """)

    # Commit and close the connection
    conn.commit()
    conn.close()
```

```python
def insert_data_from_dataframe():
    # Connect to MySQL server
    try:
        conn = mysql.connector.connect(host=host, user=user, password=password, database=database)
        cursor = conn.cursor()

        # Insert data into countries table
        countries_data = [(country.lower().strip(),) for country in df['Country'].unique()]
        cursor.executemany("INSERT INTO Countries (name) VALUES (%s)", countries_data)

        # Insert data into ExperienceLevels table
        experience_levels_data = [(level,) for level in df['Experience level'].dropna().unique()]
        cursor.executemany("INSERT INTO experience_levels (name) VALUES (%s)", experience_levels_data)
        
        # Insert data into ExperienceLevels table
        job_types_data = [(level,) for level in df['Job Type'].dropna().unique()]
        cursor.executemany("INSERT INTO job_types (name) VALUES (%s)", job_types_data)

        
        # Get unique facilities from the DataFrame
        facilities_data = list(set(facility for facilities_list in df['Facilities'] for facility in facilities_list))
        # Insert data into facilities table
        facilities_data = [(facility,) for facility in facilities_data]
        cursor.executemany("INSERT INTO facilities (name) VALUES (%s)", facilities_data)
        
        # Get unique facilities from the DataFrame
        facilities_data = list(set(facility for facilities_list in df['Facilities'] for facility in facilities_list))
        # Insert data into facilities table
        facilities_data = [(facility,) for facility in facilities_data]
        cursor.executemany("INSERT INTO facilities (name) VALUES (%s)", facilities_data)

        
         # Get unique requirements from the DataFrame
        requirements_data = list(set(requirement for requirements_list in df['Requirment of the company'] for requirement in requirements_list))
        # Insert data into requirements table
        requirements_data = [(requirement,) for requirement in requirements_data]
        cursor.executemany("INSERT INTO requirements (name) VALUES (%s)", requirements_data)
        
         # Insert data into jobs table
        jobs_data = [(row['JobTitle'], row['Job Title'], row['Location'], row['Salary USD']) for _, row in df.iterrows()]
        cursor.executemany("INSERT INTO jobs (jobTitle, jobDescription, location, salaryUSD) VALUES (%s, %s, %s, %s)", jobs_data)
        
        conn.commit()
        conn.close()
    except mysql.connector.Error as err:
        print(f"Error during data insertion: {err}")
        conn.rollback()  # Rollback changes in case of any error

    finally:
        # Close the connection
        conn.close()
```

```python
if __name__ == "__main__":
    delete_tables()
    create_tables()
    insert_data_from_dataframe()
    print("Successfully!")
```
