```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

df = pd.read_csv('./clean_data_with_the_countries2.csv', encoding='ISO-8859-1')  
df = df.rename(columns={"Requirment of the company ": "Requirment of the company"})
df
```

```python
# Use value_counts() on the "Company" column to get the count of offers for each company
offers_count = df['Company'].value_counts()

# Select the top 30 companies based on the number of offers
top_30_offers = offers_count.head(20)

# Create a bar chart using matplotlib
plt.figure(figsize=(12, 6))  # Set the figure size
plt.bar(top_30_offers.index, top_30_offers.values)

# Customize the plot
plt.xlabel('Company')  # X-axis label
plt.ylabel('Number of Offers')  # Y-axis label
plt.title('Top 30 Companies with the Highest Number of Offers')  # Plot title
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
job_title_counts = df['JobTitle'].value_counts()

# Select the top job titles based on the number of offers (let's take the top 5 for this example)
top_job_titles = job_title_counts

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_job_titles.index, top_job_titles.values)

# Customize the plot
plt.xlabel('Job Title')  # X-axis label
plt.ylabel('Number of Offers')  # Y-axis label
plt.title('Top 5 Job Titles with the Highest Number of Offers')  # Plot title
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

```python
job_type_counts = df['Job Type'].value_counts()

print(job_type_counts)

# Create a pie chart using matplotlib
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(job_type_counts, labels=job_type_counts.index, autopct='%1.1f%%', startangle=140)

# Customize the plot
plt.title('Distribution of Job Types')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.tight_layout()
plt.show()
```

```python
# Function to convert the string representation to a list
def convert_to_list(string_representation):
    return ast.literal_eval(string_representation)

# Apply the function to the column "Requirment of the company"
df["Requirment of the company"] = df["Requirment of the company"].apply(convert_to_list)
df["Facilities"] = df["Facilities"].apply(convert_to_list)
```

```python
# Combine all lists into one and convert to a Series
combined_list = df["Requirment of the company"].sum()
combined_series = pd.Series(combined_list)

# Get the value counts of each unique value in the combined Series
value_counts = combined_series.value_counts()

filtered_value_counts = value_counts[value_counts > 80]

# Create a bar chart for the filtered value counts using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(filtered_value_counts.index, filtered_value_counts.values, color='skyblue')

# Customize the plot
plt.xlabel('Requirement of the Company')
plt.ylabel('Count')
plt.title('Count of Each Requirement with Count > 50 in the "Requirment of the Company" Column')
plt.xticks(rotation=90, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Combine all lists into one and convert to a Series
combined_list = df["Facilities"].sum()
combined_series = pd.Series(combined_list)

# Get the value counts of each unique facility in the combined Series
value_counts = combined_series.value_counts()

# Filter the value counts to only include facilities with count > 50
filtered_value_counts = value_counts[value_counts > 50]

# Create a bar chart for the filtered value counts using matplotlib
plt.figure(figsize=(10, 6))
plt.bar(filtered_value_counts.index, filtered_value_counts.values, color='lightgreen')

# Customize the plot
plt.xlabel('Facility')
plt.ylabel('Count')
plt.title('Count of Each Facility with Count > 50 in the "Facilities" Column')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
salary_stats = df.groupby('JobTitle')['Salary USD'].agg(['max', 'min', 'mean'])

# Sort the DataFrame by 'max' salary in descending order
salary_stats_sorted = salary_stats.sort_values(by='max', ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(salary_stats_sorted.index, salary_stats_sorted['max'], color='blue', label='Max Salary')

# Plot the 'mean' salary
plt.bar(salary_stats_sorted.index, salary_stats_sorted['mean'], color='green', label='Mean Salary')

# Plot the 'min' salary
plt.bar(salary_stats_sorted.index, salary_stats_sorted['min'], color='red', label='Min Salary')



# Customize the plot
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')
plt.title('Salary Statistics by Job Title')
plt.xticks(rotation=90)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Assuming you have a DataFrame named 'df' with a 'Country' column and a 'Salary USD' column
# Replace 'df' with your actual DataFrame name

# Group by 'Country' and calculate the mean 'Salary USD' for each country
country_stats = df.groupby('Country')['Salary USD'].mean().reset_index()

# Create the choropleth map using Plotly
fig = px.choropleth(country_stats, 
                    locations='Country', 
                    locationmode='country names', 
                    color='Salary USD', 
                    hover_name='Country', 
                    hover_data=['Salary USD'], 
                    color_continuous_scale='YlGnBu',
                    labels={'Salary USD': 'Average Salary (USD)'}
                   )

# Customize the layout of the map
fig.update_layout(
    title_text='Average Salary (USD) by Country',
    geo=dict(showframe=False, 
             showcoastlines=False, 
             projection_type='equirectangular'
            )
)

# Show the map
fig.show()
```
